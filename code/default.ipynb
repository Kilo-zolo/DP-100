```python
# Make 2 copies of dataframe to use for 2 seperate normilization methods
df_mms = df.copy()
df_z = df.copy()
```

```python
#MinMaxScaler normilization

#import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# Create a scale variable that holds the minmaxscaler function
scale = MinMaxScaler()

# select all numerical datatypes and group them together under a var
num_col = df_mms.select_dtypes(include = 'number').columns

# for each numerical column use the variable initially created to store the scaler function 
# Transforming the selcted grouped columns and inputting it into the og df
df_mms[num_col] = scale.fit_transform(df_mms[num_col])
```

```python
#Z-score normilization

#import Z-Score
from sklearn.preprocessing import StandardScaler

# Create a standard scaler variable
scale = StandardScaler()

# select all numerical datatypes and group them together under a var
num_col2 = df_z.select_dtypes(include='number').columns

# for each numerical column use the variable initially created to store the scaler function 
# Transforming the selcted grouped columns and inputting it into the og df
df_z[num_col2] =scale.fit_transform(df_z[num_col2])
```

```python
# Grouping all object columns under var cols
col = df.select_dtypes(include = "object").columns
```

```python
# converting gropued columns under var cols from object -> categorical
df[col] = df[col].astype("category")
```

```python
# Creating copy of the dataframe
dfa = df.copy()
```

```python
# Remove all nan values from dataset 
dfa = dfa.dropna()
```

```python
# Convert categorical values to numerical categorical values
# example: male:0 female:1
# This is called label encoding

# for column in categorical grouped colums  
for i in col:
    # categorically code the columns and input into datset 
    # this inturn encodes the labels as per required
    dfa[i] = dfa[i].cat.codes
```

```python
# While label encoding helps us use the non-numerical data more and get further insights on them by converting them into numbers
# Some numerical procedures on data that has been transformed to be numerical would cause  us issues
# Example: London: 1, New Delhi: 2, California: 3 
# If 1+2= 3, does that mean London + New Delhi = California
# It does not, hence when calculating the mean and std dev of these datasets the insights might throw our model off 
# To solve against this we use one-hot encoding 
# Where the dataframe is divided into three categories and three columns are added for the three seperate categories
# Category 1: London: 1, New Delhi: 0, Caligfonia: 0
# Category 2: London: 0, New Delhi: 1, Caligfonia: 0
# Category 3: London: 0, New Delhi: 0, Caligfonia: 1
# "dummy variables" are formed to seperate these features and act as seperate columns in the dataframe
# we use "drop_first=true" to drop the first dummy variable out of each category to decrease co-linearity and model bias
dum = pd.get_dummies(df, drop_first=True)
```
